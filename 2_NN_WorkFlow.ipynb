```python
import torch
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.head()
```

```python
df.drop(columns=["id",'Unnamed: 32'],inplace=True)
```

```python
df.info()
```

```python
df.nunique()
```

#train test split

```python
#train test split
x=df.drop(columns=["diagnosis"])
y=df["diagnosis"]

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#scaling

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

#label encoding

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

#numpy to tensor

```python
x_train_tensor=torch.from_numpy(x_train.astype(np.float32))
x_test_tensor=torch.from_numpy(x_test.astype(np.float32))
y_train_tensor=torch.from_numpy(y_train.astype(np.float32))
y_test_tensor=torch.from_numpy(y_test.astype(np.float32))
```

```python
print(x_train_tensor.shape)
print(x_test_tensor.shape)
print(y_train_tensor.shape)
print(y_test_tensor.shape)
```

#model

```python
class MyModel():
  def __init__(self,x):

    self.w=torch.randn(x.shape[1],1,requires_grad=True)   #initialising weights equal to no.of cols in x
    self.b=torch.randn(1,requires_grad=True)              #initialising bias equal to no.of neurons

  def forward(self,x):
    z=torch.matmul(x,self.w)+self.b
    y_hat=torch.sigmoid(z)
    return y_hat

  def loss(self,y_hat,y):
    y=y.unsqueeze(1)
    loss=torch.nn.functional.binary_cross_entropy(y_hat,y)
    return loss

  def train(self,x,y,epochs=25,lr=0.01):

    for epoch in range(epochs):
      #forward
      y_hat=self.forward(x)

      #loss
      loss=self.loss(y_hat,y)
      print("epoch:",epoch,"loss:",loss.item())

      #backward
      loss.backward()

      #parameter update
      with torch.no_grad():           # not storing history or gradients
        self.w-=lr*self.w.grad
        self.b-=lr*self.b.grad

      #zeroing gradients
      self.w.grad.zero_()
      self.b.grad.zero_()

  def predict(self,x):
    y_pred=self.forward(x)
    y_pred=torch.where(y_pred>0.5,1,0)
    return y_pred
```

```python
#creaing instance
model=MyModel(x_train_tensor)
```

```python
#train
model.train(x_train_tensor,y_train_tensor)
```

```python
#predict
y_pred=model.predict(x_test_tensor)

#accuracy
accuracy=torch.mean((y_pred.float()==y_test_tensor.float()).float())
print("accuracy:",accuracy)
```
